```python
#creating a data frame
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
df=pd.DataFrame({"From_To":["LonDon_paris","Madrid_miLAN","londON _StockhOlm","Budapest_PaRis","Brussels_londOn"],"FlightNumber":[10045,np.nan,10065,np.nan,10085],"RecentDelays":[[23,47],[],[24,43,87],[13],[67,32]],"Airline":["KLM(!)","<Air France>(12)","(British Airways. )","12. Air France",'''Swiss Air''']})
df
```

```python
#1
df['FlightNumber']=df['FlightNumber'].interpolate().astype(int)                                
df
```

```python
#2
temporary_df = df.From_To.str.split("_",expand=True)
temporary_df.columns=['From','To']
temporary_df
```

```python
#3
temporary_df["From"]=temporary_df['From'].str.capitalize()
temporary_df["To"]=temporary_df['To'].str.capitalize()
temporary_df
```

```python
#4 
df1=df.drop(["From_To"],axis=1)
df2=df1.join(temporary_df)
df2
```

```python
#5
delays= df.apply(lambda x: pd.Series(x['RecentDelays']),axis=1).stack().reset_index(level=1, drop=True)
delays.name = 'delay_1'
df.drop(["RecentDelays"],axis=1).join(delays)

```
